```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
precip = pd.read_csv('datos_climatologicos.csv')
```

```python
precip.info()
```

```python
precip[precip['Precip_mm'] == 'NULO'] 
```

```python
precip['Precip_mm'] = precip['Precip_mm'].replace('NULO', 0)
```

```python
precip['Precip_mm'] = precip['Precip_mm'].astype(float)
precip['Fecha'] = pd.to_datetime(precip['Fecha'], format='%d/%m/%Y')
```

```python
lluvias = precip[precip['Precip_mm'] >= 50][['Fecha','Precip_mm']]
```

```python
lluvias
```

```python
lluvias_2 = lluvias.sort_values(by = 'Precip_mm', ascending = False)
```

```python
lluvias_2.head(15)
```

```python
# crear intervalos de 50 en 50
bins = range(50, 501, 50)
```

```python
labels = ['50-100', '100-150', '150-200', '200-250','250-300', '300-350', '350-400', '400-450', '450-500']
```

```python
lluvias_2['intervalos'] = pd.cut(lluvias_2['Precip_mm'], bins=bins, labels=labels, include_lowest=True)
```

```python
lluvias_2
```

```python
lluvias_2.to_csv("lluvias_2_editado.csv", index=False)
```

```python
conteo = lluvias_2['intervalos'].value_counts().sort_index()
print(conteo)
```
